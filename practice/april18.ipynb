```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
from pydataset import data
mtcars = data('mtcars')
mtcars_copy = mtcars.copy()
mtcars.head(mtcars.shape[0])
```

```python
mtcars['vs'] = mtcars_copy['vs'].map({0:'vshaped', 1:'straight'})
mtcars.head(mtcars.shape[0])
```

```python
mtcars[mtcars['cyl']>5].sort_values('cyl', ascending=True)
```

```python
mtcars['am'] = mtcars_copy['am'].map({0:'automatic', 1:'manual'})
mtcars
```

```python
mtcars.dtypes
```

```python
# how many cars have gear=5, automatic am=0
mtcars[(mtcars['gear'] == 5) & (mtcars['am'] == 'automatic')]
```

```python
mtcars[(mtcars['gear'] >= 4) & (mtcars['am'] == 'automatic')]
```

```python
# shows mean horsepower per each cylinder
mtcars.pivot_table(values='hp', index='cyl', observed=False)
```

```python
mtcars.pivot_table(values='mpg', index=['gear', 'am'], observed=False, aggfunc=['sum', 'mean'])
```

```python
# reset index makes the table iterable
mtcars.pivot_table(values='mpg', index=['gear', 'am'], observed=False, aggfunc=['sum', 'mean']).reset_index()
```

```python
# inserting data as a parameter, querying median mpg per gear
pd.pivot_table(data=mtcars, observed=False,
               index='gear',
               values='mpg',
               aggfunc='median'
              ).reset_index()
```

```python
pd.pivot_table(data=mtcars, observed=False, index='gear', columns='am', values='mpg', aggfunc=['mean', 'max', 'min' ,'count','sum','std'] ).reset_index()
```
