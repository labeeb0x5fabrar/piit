# Strings Manipulation

```python
# strings in python
string1 = 'Dhiraj'
print(string1)
```

```python
print(string1[0])
string1 = 'Upadhaya'
print(string1)
```

```python
# --------------------
# 1. Creating strings
# --------------------
print('1. Creating Strings')
s1 = "Hello"
s2 = 'World'
s3 = """This is a multiline string
Second line"""
print(s1)
print(s2)
print(s3)
```

```python
# 2. String concatenation
print("\n2. Concatenation")
full = s1 + " " + s2
print("Concatenated:", full)
```

```python
# 3. repetition
repeat ="hi how are you? " * 5
print(repeat, end="end of string") 
```

```python
# 4. indexing
text = "python is green and mean"
print("string: ", text)
print("First character: ", text[0])
print("Last character: ", text[-1])
print("Third character: ", text[2])
print("First 2 characters: ", text[:2])
print("print until last 3 character: ", text[:-3])
print("print every second character: ", text[::2])
print("inverting the string (from last to first): ", text[::-1])
print("inverted string but every second char: ", text[::-2])

# whatever you can do with lists can be done with strings
```

### Escape characters
escape characters are strings like '\n' that encode special characters
\'	    Single Quote	

\"	    Double Quote

\\	    Backslash	

\n	    New Line	

\r	    Carriage Return	

\t	    Tab	

\b	    Backspace	

\f	    Form Feed	

\ooo	Octal value	

\xhh	Hex value

```python
# 6. string methods
name = 'DATA science'
print('original', name)
print('upper: ', name.upper())
print('lower: ', name.lower())
print('strip spaces: ', name.strip())
print('replace: ', name.replace("science", "analysis"))
```

```python
# 7. checking content
sample = 'Python123'
print("is it alphabet: ", sample.isalpha())
print("is it digit: ", sample.isdigit())
print("is it alphanumeric: ", sample.isalnum())
```

```python
# 8. splitting/joining
sentence = "python is easy to learn"
print('the sentence is ', sentence)
words = sentence.split(' ')
print(words)
joined = '*'.join(words)
print("Joined: ", joined)
```

```python
# 9. formatting
name = "Taheera"
age = 25
gender = 'F'
# old format
print('My name is %s and I am %d years old, and my gender is %s' % (name, age, gender))
# format()
print('My name is {} and I am {} years old, and my gender is {}'.format(name, age, gender))
# f-string
print(f'My name is {name} and I am {age} years old, and my gender is {gender}')
```

```python
# strings are immutable
text = "hello"
text = 'H' + text[1:]
print("modified: ", text)
```

# More (using dataframes)

```python
import pandas as pd
from pydataset import data
```

```python
# Create mtcars dataset with car names
data = {
    "car_name": [ "Mazda RX4", "Mazda RX4    Wag", "Datsun 710","Hornet 4 Drive", "Hornet   Sportabout  ", "Valiant", "Duster 360", "merc 240D", "Merc 230", "Merc 280" ],
    "mpg": [21.0, 21.0, 22.8, 21.4, 18.7, 18.1, 14.3, 24.4, 22.8, 19.2]}
df = pd.DataFrame(data)
print("Original Data:\n", df)
```

```python
# Search
mazda_cars = df[df['car_name'].str.contains("Mazda")]
print("Mazda cars: \n")
mazda_cars
```

```python
# showing case parameter to switch case sensitivity
df[df['car_name'].str.contains("Merc", case=True)]
```

```python
df[df['car_name'].str.contains("merc", case=False)]
```

```python
# cars starting with merc
merc_cars = df[df['car_name'].str.startswith("Merc")]
print("Cars starting with 'Merc':\n", merc_cars)
```

```python
# here you can also put a tuple (only accepts tuple (no lists))
df[df['car_name'].str.startswith(('merc', 'Merc'))]
```

```python
df[df['car_name'].str.endswith('Drive')]
```

```python
# Replace strings
# making new column for edited names
df['car_name_updated'] = df['car_name'].str.replace('Merc', 'Mercedes', regex=False)
print("Updated Names:\n", df[['car_name', 'car_name_updated']])
```

```python
# putting 'case=False'
df['car_name_updated'] = df['car_name'].str.replace('Merc', 'Mercedes', case=False, regex=False)
print("Updated Names:\n", df[['car_name', 'car_name_updated']])
```

```python
# splitting brand names to extract categories, useful for feature engineering
df['brand'] = df['car_name'].str.split().str[0]
print("Extracted Brand:\n", df[['car_name', 'brand']])
```

```python
df['upper_name'] = df['car_name'].str.upper()
df['lower_name'] = df['car_name'].str.lower()
df
```

```python
# clean names by trimming spaces
df['clean_name'] = df['car_name'].str.strip()
print("Clean names:\n", df[['car_name', 'clean_name']])
```
