# Graphing

```python
import pandas as pd
import numpy as np
from pydataset import data
import matplotlib.pyplot as plt
```

```python
df = data('mtcars')
df.head(df.shape[0])
```

```python
# histogram
# kind = ('line', 'bar', 'barh', 'kde', 'density', 'area', 'hist', 'box', 'pie', 'scatter', 'hexbin')
df['mpg'].plot(kind='hist', title='Distribution of MPG')
```

```python
# boxplot
df.boxplot(column='mpg', by='cyl')
plt.suptitle('')
plt.show()
```

```python
# line plot
df.sort_values('wt').plot(x='wt', y='mpg', kind='line', title='Weight vs MPG')
plt.show()
```

# When to use
- Feature Relationship → scatter plot = df.corr() + heatmap
- Understanding Distribution → histogram = df.plot(kind='density')
- Categorical Comparison → bar/box plot = df.plot(kind='bar')

```python
df.plot(kind='bar')
```

```python
df.plot(kind='density')
```

```python
df.boxplot(column='hp', by='gear')
plt.suptitle('')
plt.show()
```

```python
df['Horsepower'] = pd.cut(df['hp'], bins=[0, 100, 180, 250, df['hp'].max()], labels=['Very Low', 'Low', 'Medium', 'High'], include_lowest=True)
df.boxplot(column='wt', by='Horsepower')
plt.title("Weight vs Horsepower")
plt.ylabel("Weight")
plt.show()
```

```python
df["Transmission"] = df['am'].map({0:"automatic", 1:"manual"})
df['wt'].plot(kind='hist', title='Distribution of weight')
```

```python
plt.scatter(y=df['mpg'], x=df['wt'])
plt.suptitle("Mpg vs Weight")
plt.xlabel("Weight")
plt.ylabel("Mpg")
plt.show()
```
