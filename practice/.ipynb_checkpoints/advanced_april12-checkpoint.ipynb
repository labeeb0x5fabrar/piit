```python
# import
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
from pydataset import data
```

```python
# obtaining cars data
from pydataset import data
mtcars = data('mtcars')
mtcars.head()
```

```python
# how many entries that contain each unique value
mtcars["gear"].value_counts()
```

```python
# making a fresh copy for backup
mtcars_edit = mtcars.copy()
```

```python
mtcars_edit["am"] = mtcars_edit["am"].map({0:"auto", 1:"manual"})
mtcars_edit.head()
```

```python
mtcars["hp"].sort_values()
```

```python
mtcars_hp = mtcars.copy()
bins = [0, 100, 180, 250, mtcars['hp'].max()]
labels = ['Very Low', 'Low', 'Medium', 'High']
mtcars_hp['hp level'] = pd.cut(mtcars['hp'], bins=bins, labels=labels, include_lowest=True)
```

```python
mtcars_hp.head(mtcars.shape[0])
```

```python
mtcars_cat = mtcars_hp.copy()
mtcars_cat = mtcars_cat.astype({'cyl':'category', 'vs':'category', 'carb':'category', 'am':'category', 'gear':'category'})
mtcars_cat.head(mtcars.shape[0])
```
