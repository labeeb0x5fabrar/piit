```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
#Create Data
rawdata = [
    [1, 'Taheera', 70, 86],
    [2, 'Aelita', 56, 90],
    [3, 'Khulan', 87, 90],
    [4, 'Batgerel', 71, 79],
    [5, 'Manisha', 88, 76],
]

df = pd.DataFrame(rawdata, columns=['Rollno','Name', 'Maths', 'English'])
df.head()
```

# Wide to Long

```python
df_long = pd.melt(df, id_vars=['Name'], value_vars=['Maths', 'English'], var_name='Subject', value_name='Marks')
df_long
```

```python
df_long.groupby('Subject')['Marks'].agg(['sum', 'mean', 'count'])
```
