## Functions in Pandas
- Understanding Data
    - type  # type of object
    - shape # rows & cols in data
    - columns # column names
    - dtypes  # datatype 
    - info()  # datatypes & non-null counts
    - sample(x) # randon x rows
    - head(x) # first x rows
    - tail(x) # last x rows
- Summary Statistics
    - describe()  #summary stats for numeric cols
    - describe(include='all')  #summary stats for all cols
    - mtcars.nunique() #count of unique values per column
    - value_counts()  # freq count for a series
- Missing/Duplicates
    -  isnull().sum()   # missing value count per cols
    -  notnull().sum()  # non missing value count
    -  duplicated(().sum() # count of duplicate rows
- View Specific Col/ Row
    -  ['col_name']   # single column
    -  [['col1','col2']] # multiple columns
    -  iloc[0]   #1st row by index position
    -  loc[0] #1st row by label
    -  iloc[0:5]   # first 5 rows
-  Basic Data Insights
    -  corr(numeric_only=True)  #correlation matrix (numeric only)
    -  mean(numeric_only=True)   #mean of numeric cols
    -  mode().iloc(0) #most frequent values
    -  median(numeric_only=True) # median of numeric cols

# Check for Duplicates
- duplicated()  #check for duplicate rows
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html

```python
#?mtcars.duplicated
mtcars = data('mtcars')
mtcars.head(2)
```

```python
mtcars2 = mtcars.copy().reset_index().rename(columns={'index':'car'})  #duplicate, reset_index (rownames to column by index), rename by index to car
mtcars2.head()
```

```python
mtcars2.shape   #car - 12
```

### simulate duplicate rows

```python
mtcars2[mtcars2['car'] == 'Mazda RX4']  #all cars with name of Mazda RX4
```

```python
mazda2 = mtcars2[mtcars2['car'] == 'Mazda RX4']
datson2 = mtcars2[mtcars2['car'] == 'Datsun 710']
# copied mazda and datson car in variable/ row 
```

```python
mtcars_dup = pd.concat([mtcars2, mazda2, datson2, mazda2], ignore_index=True) # 2 more cars Mazda, 1 more car for Datson
mtcars_dup #.sort_values('car')
```

```python
mtcars_dup.shape
```

```python
mtcars_dup[mtcars_dup['car'] == 'Mazda RX4']
#duplicate cars - 3 cars of Mazda RX4
```

```python
mtcars_dup[mtcars_dup['car'] == 'Datsun 710']
```

```python
mtcars_dup.duplicated().sum()  #duplicate rows - 3
```

```python
#mtcars_dup.duplicated(subset = ['gear']).sum()
```

```python
#mtcars_dup.duplicated(keep=False)  #mark all duplicates
```

```python
mtcars_dup.drop_duplicates()
```

```python
#?mtcars_dup.drop_duplicates
```

```python
mtcars_dup.drop_duplicates(keep='first')  #first
```

```python
mtcars_dup.drop_duplicates(keep='last')
```

```python
mtcars_dup.drop_duplicates(keep=False)  #remove duplicates
```

```python
mtcars_dup[mtcars_dup.duplicated(subset = ['vs','am','gear','carb'])].sort_values(['vs','am','gear','carb'])
```

```python
mtcars[mtcars.duplicated(subset = ['gear','am','cyl','carb'], keep=False)].sort_values(by=['gear','am','cyl','carb'])
#keep=False marks all occurrences of duplicates, not just the later ones.
```

```python
mtcars_dup[~ mtcars_dup.duplicated(subset = ['gear','am','cyl','carb'], keep=False)].sort_values(by=['gear','am','cyl','carb'])
# non duplicates wrt column in subjset
```

```python
mtcars_dup.drop_duplicates(subset = ['gear','am','cyl','carb'], inplace=False)
#remove duplicates by selected columns
```
